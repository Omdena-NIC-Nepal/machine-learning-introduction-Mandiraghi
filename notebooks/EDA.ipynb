# Exploratory Data Analysis

```python
#Load Libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import scipy
from scipy import stats
```

```python
#Load Datasets
boston_housing = pd.read_csv("/Users/mandiraghimire/Desktop/MghiGitManu/machine-learning-introduction-Mandiraghi/BostonHousing.csv")
```

```python
print ("Boston Housing Statistics:")
print(boston_housing.describe())
```

## Identify missing values and outliers.

```python
print("Missing Values: ")
#Missing values 
print(boston_housing.isnull().sum())
```

```python
# Checking Outliers using boxplot 
for col in boston_housing.columns: 
    plt.figure(figsize=(6, 5))
    sns.boxplot(x = boston_housing[col])
    plt.title(f"Boxplot for {col}")
    plt.show()
```

## Visualize relationships between features and the target variable.

```python
#Establish a target variables and features 
y = boston_housing['medv']
x = boston_housing.drop('medv', axis = 1)
```

```python
# Viusalize the features and target variable and understand the datasets 
plt.figure(figsize=(12, 8))
sns.heatmap(boston_housing.corr(), annot = True, cmap='coolwarm', fmt = '.2f')
plt.title("Correlation Matrix - Boston Housing")
plt.show()
```

```python
#Pairplot of Most Correlated Features 
Corr_boston_housing = boston_housing.corr()['medv'].abs().sort_values(ascending= False).index[1:6]
sns.pairplot(boston_housing[Corr_boston_housing.to_list() + ['medv']])
plt.suptitle("Most Features vs MEDV",y = 1.02 )
plt.show()
```
